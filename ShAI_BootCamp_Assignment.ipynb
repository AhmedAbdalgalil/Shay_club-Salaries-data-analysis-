 <h1 style="color:red;">Simple Correlation Analysis: : 🌟</h1>
Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.

```python
# Making data frame from the csv file
df_corr = df[["BasePay","OvertimePay","OtherPay","Benefits","TotalPay","TotalPayBenefits"]]
 
# To find the correlation among
# the columns using kendall method
df_corr.corr(method='kendall')
```

#### Here are some comments on the  correlation matrix:

#### High Correlations:

The diagonal elements of the matrix are all 1.0, which is expected as they represent the correlation of each variable with itself.
Notably, the correlation between "TotalPay" and "TotalPayBenefits" is very high (0.888485), which is expected since "TotalPayBenefits" includes "TotalPay" as one of its components.
#### Strong Positive Correlations:

"BasePay" has a strong positive correlation with "TotalPay" (0.887760) and "TotalPayBenefits" (0.832695). This indicates that as "BasePay" increases, both "TotalPay" and "TotalPayBenefits" tend to increase as well.

There is also a strong positive correlation between "TotalPay" and "TotalPayBenefits" (0.888485), which makes sense since "TotalPayBenefits" is the sum of "TotalPay" and "Benefits."

#### Moderate Positive Correlations:

"OvertimePay" and "OtherPay" show moderate positive correlations with "TotalPay" and "TotalPayBenefits," indicating that higher overtime pay or other pay is associated with higher total pay and total pay with benefits.

"Benefits" has a moderate positive correlation with "TotalPayBenefits" (0.595304), indicating that employees with higher benefits tend to have higher total pay with benefits.

#### Weak Positive Correlations:

The correlation between "OvertimePay" and "Benefits" is relatively weak (0.169444).
Directionality of Correlations:

#### All the correlations are positive, indicating that as one variable increases, the others tend to increase as well. This aligns with expectations, as higher pay components are generally associated with higher total pay and total pay with benefits.
